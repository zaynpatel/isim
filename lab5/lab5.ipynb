```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('heartbeat')
df
```

```python
data = df.drop(['t1', 'ch1'], axis =1)
data
```

### EKG Trace Graph

**Caption:** A graph of heartbeat changing over time where the Voltage is measuring the output electricity from the left and right wrist and the back of hand. 

```python
plt.figure().set_figwidth(25)
plt.plot(data['t2'], data['ch2'])
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('EKG Trace Graph')
```
